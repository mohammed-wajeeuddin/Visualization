```python
import geopandas as gpd
import matplotlib.pyplot as plt
import os

# Define your data directory
data_dir = "/content/drive/MyDrive/Freelancing/Data Science Assignment - Monore Lake/Lake_monroe_boundary_and_network"

# Load different versions of shapefiles
shapefiles = {
    "EPSG 4269": "LakeMonroe_NHDFlowline_epsg4269/LakeMonroe_NHDFlowline_epsg4269.shp",
    "EPSG 4326": "LakeMonroe_NHDFlowline_epsg4326/LakeMonroe_NHDFlowline_epsg4326.shp",
    "EPSG 3087": "LakeMonroe_WBDHU10_epsg3087/LakeMonroe_WBDHU10_epsg3087.shp"
}

colors = ["blue", "green", "red"]

# Create subplots with 1 row and as many columns as shapefiles
fig, axes = plt.subplots(
    nrows=1,
    ncols=len(shapefiles),
    figsize=(18, 6),  # wider figure for side-by-side
    sharex=False,
    sharey=False
)

# Loop through each shapefile and plot on a separate axis
for (name, rel_path), ax, color in zip(shapefiles.items(), axes, colors):
    full_path = os.path.join(data_dir, rel_path)
    if os.path.exists(full_path):
        gdf = gpd.read_file(full_path)

        # Plot
        gdf.plot(ax=ax, color=color, edgecolor="black", alpha=0.5)
        ax.set_title(name, fontsize=14)
        ax.set_xlabel("Longitude", fontsize=12)
        ax.set_ylabel("Latitude", fontsize=12)
        ax.grid(True)

plt.suptitle("Lake Monroe in Different Coordinate Reference Systems", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
target_crs = "EPSG:4326"

# Prepare a figure
fig, ax = plt.subplots(figsize=(10, 8))

# Plot watershed layers first
for i, (label, rel_path) in enumerate(watershed_shapefiles.items()):
    full_path = os.path.join(data_dir, rel_path)
    if os.path.exists(full_path):
        gdf = gpd.read_file(full_path)
        # Reproject if needed
        if gdf.crs and gdf.crs.to_string() != target_crs:
            gdf = gdf.to_crs(target_crs)
        gdf.plot(ax=ax, alpha=0.4, edgecolor="black", label=label if i == 0 else None)

# Plot flowlines on top
for i, (label, rel_path) in enumerate(flowline_shapefiles.items()):
    full_path = os.path.join(data_dir, rel_path)
    if os.path.exists(full_path):
        gdf = gpd.read_file(full_path)
        # Reproject if needed
        if gdf.crs and gdf.crs.to_string() != target_crs:
            gdf = gdf.to_crs(target_crs)
        gdf.plot(ax=ax, color="blue", alpha=0.8, linewidth=1, label=label if i == 0 else None)

plt.title("Lake Monroe Watershed & Flowlines (Overlay, EPSG:4326)", fontsize=15)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Path to the 'dissolved v2' watershed shapefile
wdissolved_shp = os.path.join(data_dir,
                              "LakeMonroe_WBDHU10_dissolved_v2",
                              "LakeMonroe_WBDHU10_dissolved_v2.shp")

# Load the shapefile (this is polygons for the watershed boundary)
gdf = gpd.read_file(wdissolved_shp)

# Reproject to a suitable projected CRS for accurate area calculations
# Let's use EPSG:3087 since you already have a shapefile in that projection
target_area_crs = "EPSG:3087"
if gdf.crs and gdf.crs.to_string() != target_area_crs:
    gdf = gdf.to_crs(target_area_crs)

# Compute the total area in sq meters
lake_area_m2 = gdf.geometry.area.sum()
print("Total watershed area (sq. meters):", lake_area_m2)
```

```python
import numpy as np
# Let's define a list of years
years = [2015, 2017, 2019, 2021, 2023, 2025]

# We'll store data in a list of dicts
time_series_data = []
baseline_area = lake_area_m2  # from above
np.random.seed(42)  # for reproducibility

for yr in years:
    # Add some random fluctuation to the area:
    # e.g., plus/minus up to 2% of the baseline
    fluctuation = baseline_area * (np.random.uniform(-0.02, 0.02))
    new_area = baseline_area + fluctuation

    time_series_data.append({"Year": yr, "Area_sq_m": new_area})

df_ts = pd.DataFrame(time_series_data)
df_ts = df_ts.sort_values("Year")  # ensure ascending order
df_ts
```

```python
plt.figure(figsize=(8, 5))
plt.plot(df_ts["Year"], df_ts["Area_sq_m"], marker="o", linestyle="-", color="blue")

# Label points
for i, row in df_ts.iterrows():
    plt.text(row["Year"], row["Area_sq_m"],
             f"{row['Area_sq_m']:.0f}",
             ha='center', va='bottom', fontsize=9)

plt.title("Simulated Watershed Area Changes (2015 - 2025)")
plt.xlabel("Year")
plt.ylabel("Area (sq. meters)")
plt.grid(True)
plt.show()
```

```python

# 1) Create a monthly date range (for 10 years)
date_range = pd.date_range(start="2015-01-01", periods=120, freq="M")

# 2) Simulate Water Level (m)
#    - Baseline around 2.5 m, with a slight seasonal sine wave + random noise
np.random.seed(42)  # For reproducible randomness
water_level = 2.5 + 0.3*np.sin(np.linspace(0, 4*np.pi, 120)) + np.random.normal(0, 0.05, 120)

# 3) Simulate Precipitation (mm/month)
#    - Varies seasonally, let's say baseline 100 mm with a strong sinusoidal pattern + noise
precipitation = 100 + 30*np.sin(np.linspace(0, 2*np.pi, 120)) + np.random.normal(0, 5, 120)

# 4) Simulate Temperature (°C)
#    - Let's assume a typical annual cycle: average 20°C, amplitude 10°C, plus some noise
temperature = 20 + 10*np.sin(np.linspace(0, 12*np.pi, 120)/6) + np.random.normal(0, 1, 120)

# 5) Combine into a single DataFrame
df = pd.DataFrame({
    "Date": date_range,
    "Water_Level_m": water_level,
    "Precipitation_mm": precipitation,
    "Temperature_C": temperature
})
df.set_index("Date", inplace=True)
df.head()
```

```python
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot Water Level
color1 = "blue"
ax1.set_xlabel("Date")
ax1.set_ylabel("Water Level (m)", color=color1)
ax1.plot(df.index, df["Water_Level_m"], color=color1, label="Water Level")
ax1.tick_params(axis='y', labelcolor=color1)

# Create a second y-axis for precipitation
ax2 = ax1.twinx()
color2 = "green"
ax2.set_ylabel("Precipitation (mm/month)", color=color2)
ax2.plot(df.index, df["Precipitation_mm"], color=color2, label="Precipitation")
ax2.tick_params(axis='y', labelcolor=color2)

# For a third parameter, you can create a new axis or overlay differently
# Let's overlay Temperature on ax1 but scale it differently
ax3 = ax1.twinx()  # This shares the x-axis with ax1
color3 = "red"
# Move the spines to avoid overlap
ax3.spines["right"].set_position(("outward", 60))
ax3.set_ylabel("Temperature (°C)", color=color3)
ax3.plot(df.index, df["Temperature_C"], color=color3, linestyle="--", label="Temperature")
ax3.tick_params(axis='y', labelcolor=color3)

fig.tight_layout()
plt.title("Lake Monroe - Multiple Parameters Over Time")
plt.show()
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(10, 8), sharex=True)

# 1) Water Level
axes[0].plot(df.index, df["Water_Level_m"], color="blue")
axes[0].set_title("Water Level (m)")
axes[0].grid(True)

# 2) Precipitation
axes[1].plot(df.index, df["Precipitation_mm"], color="green")
axes[1].set_title("Precipitation (mm/month)")
axes[1].grid(True)

# 3) Temperature
axes[2].plot(df.index, df["Temperature_C"], color="red")
axes[2].set_title("Temperature (°C)")
axes[2].grid(True)

plt.suptitle("Lake Monroe - Multiple Parameter Time Series", fontsize=14)
plt.xlabel("Date")
plt.tight_layout()
plt.show()
```
